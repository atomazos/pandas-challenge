```python
#Import dependencies
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
csv_path = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(csv_path)
```

```python
info_pd = pd.DataFrame(purchase_data_df, columns=['Purchase ID', 'SN', 'Age', 'Gender', 'Item ID', 'Item Name', 'Price'])
info_pd
```

```python
#A.TOTAL NUMBER OF PLAYERS
total_players_pd = info_pd.drop_duplicates(subset='SN')
total_players2_pd = total_players_pd['SN'].count()
total_players2_pd
```

```python
# summary_table_1 = pd.Dataframe({"Total Number of Players": [total_players2_pd]})
# summary_table_1    
```

```python
new_info_pd = pd.DataFrame(purchase_data_df, columns=['Purchase ID', 'Item Name', 'Price'])
new_info_pd
```

```python
#B. PURCHASING ANALYSIS(TOTAL)
# 2. Average Purchase Price

avg_purchase_price = new_info_pd['Price'].mean()
avg_purchase_price
```

```python
#3. Total Number of Purchase
total_purchases_pd = new_info_pd["Purchase ID"].count()
total_purchases_pd
```

```python
#4. Total Revenue
total_revenue_pd =(total_purchases_pd * avg_purchase_price)
print(total_revenue_pd)
```

```python
# 1. Number of Unique Items
number_items_pd = len(new_info_pd["Item Name"].unique())
number_items_pd
```

```python
summary_table_2 = pd.DataFrame({"Number of Unique Items": [number_items_pd],
                              "AVG Purchase Price": [avg_purchase_price],
                              "Total Purchases": [total_purchases_pd],
                              "Total Revenue": [total_revenue_pd]})
print('Purchasing Analysis')
print('_______________________________________________________________________________')

print(summary_table_2)
```
